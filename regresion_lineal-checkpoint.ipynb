# *1. Calentamiento*
>El objetivo de la Regresión Lineal es obtener un modelo que permita establecer la relación de dependencia entre una variable $Y$ conocida como *variable dependiente* o *variable respuesta* y un conjunto de variables *independientes* o *explicativas* $X_i$ ; $i=1,...,n$.
En el ejemplo más sencillo del que trata este resumen, conocido como **Regresión Lineal Simple**, se tiene una sóla variable independiente $X$ como explicativa de los valores que toma la variable dependiente $Y$.

>![Regresión Lineal Simple](https://upload.wikimedia.org/wikipedia/commons/3/3a/Linear_regression.svg)

>*Gráfico de una Regresión Lineal Simple, es decir, una sóla variable predictora $X$ y una variable respuesta $Y$*


# *2. Algo de historia*
>El método de estimación de Mínimos Cuadrados o [MCO](https://es.wikipedia.org/wiki/M%C3%ADnimos_cuadrados_ordinarios) fue la primera Regresión Lineal documentada de la historia, publicada por el destacado matemático francés *Adrien-Marie Legendre* hacía la primera década del siglo XIX.
El término regresión se acuñó a partir de un experimento realizado a finales del siglo XIX, para ver el grado de dependencia de la estatura de los hijos según la estatura de los padres. En dicho experimento, la evidencia empírica sugirió que los hijos de padres altos tienden a ser altos y los hijos de padres bajos tienden de a ser bajos, pero en ambos casos, con una tendencia a "regresar" al promedio de las estaturas, o acercarse a la media de los datos.

# *3. La parte formal y elegante del asunto*
>El modelo de regresión lineal busca relacionar una variable aleatoria dependiente $Y$ con $n$ variables aleatorias explicativas $X_i$ con $i=1,...,n$ de tal forma que
$$
Y=\beta_0+\beta_1X_1+...+\beta_nX_n+\epsilon=\beta_0+\sum_{i=1}^{n}\beta_iX_i+\epsilon
$$
Dónde $\beta_i $ son los parámetros desconocidos de la regresión y $\epsilon$ es el residuo, que en realidad es una varible aleatoria que representa todas aquellas partes del fenómeno en estudio que no son explicadas por las variables independientes de la regresión.

>La regresión busca encontrar una estimación para los parámetros desconocidos $\beta_i$, que se notan como $\hat\beta_i$ y se conocen como *coeficientes de regresión*. Para lograr su cometido, se debe tener un conjunto de $n$ observaciones $y_i$ ; $i=1,...,n$ con sus correspondientes valores de las variables explicativas $x_{ij}$ ; $j=1,...,m$.

>Como los coeficientes de regresión son estimaciones y no hay manera de asegurar que sean exactamente iguales a los parámetros de la regresión, se tiene que
$$
Y_i=\hat\beta_0+\sum_{i=1}^{n}\hat\beta_iX_{ij}+\hat\epsilon_i
$$
donde $\hat\epsilon_i$ corresponden a la estimación de los errores de la regresión.

># Hipótesis del modelo de Regresión Lineal Clásico
>>Hay algunas hipótesis que se deben verificar para que la aplicación de un modelo de regresión lineal sea correcta:
>>> 1. El valor esperado del error es cero. Tiene sentido que siendo la regresión lineal el mejor modelo para ajustar los datos, los errores de dicho modelo tengan media a cero.
>>> 2. Homocedasticidad, es decir, la varianza de los errores existe y es finita. Se denota como $\sigma^2$.
>>> 3. Los errores tienen una distribución normal estándar, o lo que es lo mismo, siguen una función de distribución normal con media 0 y varrianza $\sigma^2$.
>>> 4. Los errores son independientes. En este sentido se dice que la covarianza entre los errores de estimación entre una observación $i$ y una observación $j$ es cero. De una manera más sencilla, el error que se comete al pronosticar un valor dado $\hat y_i$ no influye en el error que se comete al pronosticar otro valor cualquiera $\hat y_j$.
>>> 5. Los coeficientes de regresión son linealmente independientes.
>>> 6. Se supone que no hay errores de medida de las variables independientes ni errores de especificación del modelo, es decir, que el modelo está correctamente escogido para describir el fenómeno que se está estudiando.

># El caso más sencillo: Regresión Lineal Simple
>> En este caso de regresión lineal se tiene una sola variable dependiente $Y$ y una sola variable explicativa $X$, lo que lleva al siguiente enunciado de la regresión:
$$
Y=\beta_0+\beta_1X+\epsilon
$$
Donde $\beta_0$ y $\beta_1$ son los parámetros desconcidos que se van a estimar mediante los datos de la muestra escogida, y se conocen como *coeficientes de regresión*. Para la estimación, se utilizará el método de mínimos cuadrados, que en resumen lo que intenta es encontrar los $\hat \beta_i$ que minimicen la suma de los cuadrados de las distancias entre cada observación $y_i$ y su valor predicho según el modelo $\hat y_i$.

>>Sea la función de error cuadrático medio dada por
$$
S(\beta_0\beta_1)=\sum_{i=1}^{n}(y_i-\hat y_i)^2
$$
>> Esta función alcanza su mínimo en el punto $(\hat \beta_0,\hat \beta_1)$; al derivar parcialmente la función de error cuadrático medio con respecto a $\hat \beta_0$ y $\hat \beta_1$ e igualando a cero, se obtiene un sistema de dos ecuaciones con dos incógnitas (conocido como *ecuaciones normales*). Las dos soluciones de dicho sistema, son los estimadores de los paramétros de la regresión:
$$
\hat \beta_1=\frac{\displaystyle\sum_{i=1}^{n}x_iy_i-n\bar{x}\bar{y}}{\displaystyle\sum_{i=1}^{n}x_i^2-\frac{1}{n}\left(\displaystyle\sum_{i=1}^{n}x_i\right)^2}
$$

$$
\hat \beta_0=\frac{\displaystyle\sum_{i=1}^{n}x_i^2\displaystyle\sum_{i=1}^{n}y_i-\displaystyle\sum_{i=1}^{n}x_iy_i\displaystyle\sum_{i=1}^{n}x_i}{n\displaystyle\sum_{i=1}^{n}x_i^2-\left(\displaystyle\sum_{i=1}^{n}x_i\right)^2}=\bar y-\hat \beta_1\bar x 
$$

>> Finalmente, el modelo *ajustado* de Reresión Lineal Simple, es decir la ecuación que nos permite pronosticar valores de la variable $Y$ a partir de valores observados de la variable $X$ se escribe así:

$$
{\hat y=\hat \beta_0+\hat \beta_1x}
$$

> Vamos a ejecutar una regresión lineal simple con variable dependiente $Y$ *longitud del pie* y variable independiente $X$ *estatura*. No vamos a hacer diferenciación entre hombres y mujeres para este ejercicio.

```python
# Correlación lineal entre las dos variables
# ==============================================================================
corr_test = pearsonr(x = medifis.ESTATURA, y =  medifis.LONG_PIE)
print("Coeficiente de correlación = ", corr_test[0])
print("P-valor = ", corr_test[1])
```

> Corroboramos lo que ya intuíamos gráficamente, es decir, que la correlación existente entre la dos variables es altísima (de 92.78%). En ese orden de ideas, esperaríamos que la estatura del individuo sea una buena predictora de la longitud de su pie y que, a mayor estatura, se corresponda un mayor tamaño de la extremidad.

```python
X = medifis.ESTATURA
y = medifis.LONG_PIE

X=X.values.reshape(-1,1)
y=y.values.reshape(-1,1)


# Creación del modelo
# ==============================================================================
modelo = LinearRegression()
modelo.fit(X , y)
```

```python
# Información del modelo
# ==============================================================================
print("Intercepto = ", modelo.intercept_)
print("Coeficiente de regresión = ", modelo.coef_.flatten())
print("Coeficiente de determinación R^2:", modelo.score(X, y))
```

> El coeficiente de determinación $R^2$ describe la proporción de la variabilidad de la variable respuesta, en este caso longitud del pie, que es explicada por la variable independiete, estatura. En la medida en que el coeficiente esté cercano a $1$ significa que la variable predictora es buena para pronosticar los valores que puede tomar la variable dependiente.

> Así, la estatura es un muy buen predictor de la longitud del pie, dado que encontramos $R^2=86.60\%$

> El intercepto o $\hat \beta_0$ es el valor que tomaría la variable respuesta en ausencia de la variable explicativa. En este caso por ser negativo debe tratarse con cuidado, dado que en ausencia de estatura se esperaria ausencia de longitud del pie y no que la longitud se vuelva "negativa" (es decir que el pie implosione?). 

> Es mejor utilizar el paquete statsmodels porque permite revisar las hipotésis de significancia de los coeficientes de regresión, así como algunas otras pruebas de normalidad de los errores que se deben verificar en el modelo:

```python
# Creación del modelo utilizando matrices como en scikitlearn
# ==============================================================================
# A la matriz de predictores se le tiene que añadir una columna de 1s para el intercept del modelo
X = sm.add_constant(X, prepend=True)
modelo = sm.OLS(endog=y, exog=X,)
modelo = modelo.fit()
print(modelo.summary())
```

> Cómo ya nos sugería el sentido común, el intercepto resulta no ser significativo al no rechazar la hipotesis nula de significancia del parámetro con un p-valor alto de 0.17. Por tal motivo, ajustamos una nueva regresión sin intercepto:

```python
# Creación del modelo utilizando matrices como en scikitlearn y quitando el intercepto
# ==============================================================================
X = medifis.ESTATURA
X=X.values.reshape(-1,1)
modelo = sm.OLS(endog=y, exog=X,)
modelo = modelo.fit()
print(modelo.summary())
```

> Se verifica la significancia del coeficiente de regresión,  porque el p-valor pequeño de la prueba indica que el parámetro es no-nulo.

> El coeficiente de regresión o $\hat \beta_1$ es el que nos indica cómo varía el crecimiento del pie a medida que varía la estatura, es decir, un crecimiento de 10 puntos porcentuales en la estatura del individuo equivale a un crecimiento de 2.3 puntos porcentuales en la longitud del pie.

```python
# Intervalos de confianza para los coeficientes del modelo
# ==============================================================================
modelo.conf_int(alpha=0.05)
```

>Con un nivel de confianza del 95%, la estimación del parámetro se encuentra en el intervalo $[0.2284 ; 0.2336]$

> Finalmente, es interesante dibujar la región de pronóstico del fenómeno, es decir, la recta de regresión con su área sombreada delimitada por el intervalo de confianza del parámetro estimado:

```python
# Predicciones con intervalo de confianza del 95%
# ==============================================================================
predicciones = modelo.get_prediction(exog = X).summary_frame(alpha=0.05)
predicciones['x'] = X
predicciones['y'] = y
predicciones = predicciones.sort_values('x')

# Gráfico del modelo
# ==============================================================================
fig, ax = plt.subplots(figsize=(6, 3.84))

ax.scatter(predicciones['x'], predicciones['y'], marker='o', color = "gray")
ax.plot(predicciones['x'], predicciones["mean"], linestyle='-', label="Recta de regresión")
ax.plot(predicciones['x'], predicciones["mean_ci_lower"], linestyle='--', color='red', label="95% Intervalo de confianza")
ax.plot(predicciones['x'], predicciones["mean_ci_upper"], linestyle='--', color='red')
ax.fill_between(predicciones['x'], predicciones["mean_ci_lower"], predicciones["mean_ci_upper"], alpha=0.1)
plt.xlabel("Estatura")
plt.ylabel("Longitud del pie")
ax.legend();
```
